```python
import pandas as pd
```

```python
df_news = pd.read_csv("data/2015_Company.csv", encoding="utf-8")
df_news
```

```python
df_news.shape
```

```python
# column name rename "編號":"id", "風險類型": "label", "段落摘要": "text"
df_news = df_news.rename(columns={"編號":"id", "風險類型": "label", "段落摘要": "text"})
df_news.head(2)
```

```python
# df_news['風險事件區間'] = pd.to_datetime(df_news['風險事件區間'])
df_news['text'] = df_news['text'].astype(str).str.replace('\t' , ' ')   # tab都換成空格
```

```python
# 只取特定欄位
df_news2 = df_news.loc[:, ['id', 'label', 'text']]
df_news2.head(2)
```

```python
# 字數小於512-2 因為還有CLS SEP
MAX_LENGTH = 510
# df_news2 = df_news2[~(df_news2["text"].apply(lambda x : len(x)) > MAX_LENGTH)]
df_news2 = df_news2[df_news2["text"].apply(lambda x : len(x) <= MAX_LENGTH) ]
df_news2 = df_news2.reset_index(drop=True)      # 重設index
df_news2['id'] = df_news2.index                 # index 變成 id column
df_news2
```

```python
df_news2['label2'] = 0
df_news2
```

```python
df_news2["label"].unique() 
```

```python
df_news2.loc[df_news2['label'] == "財務營運風險", 'label2'] = 0
df_news2.loc[df_news2['label'] == "法律風險", 'label2'] = 1
df_news2.loc[df_news2['label'] == "資安風險", 'label2'] = 2
df_news2.loc[df_news2['label'] == "勞動風險", 'label2'] = 3
df_news2.loc[df_news2['label'] == "管理風險", 'label2'] = 4
df_news2
```

```python
df_news2["label2"].unique() 
```

```python
df_news2["label"] = df_news2["label2"]
df_news2["label"] = df_news2["label"].astype(int)
df_news2 = df_news2.drop(columns=['label2'])
df_news2 = df_news2.drop(columns=['id'])
df_news2['text'] = df_news2['text'].astype(str).replace(r'\n', ' ', regex=True)
# df_news2['alpha'] = "a"
df_news2 = df_news2.rename(columns={"text": "text","label": "labels"})
df_news2
```
