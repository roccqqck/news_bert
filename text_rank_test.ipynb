```python
df_news.loc[df_news['label'] == "財務營運風險", 'label2'] = 0
df_news.loc[df_news['label'] == "法律風險", 'label2'] = 1
df_news.loc[df_news['label'] == "資安風險", 'label2'] = 2
df_news.loc[df_news['label'] == "勞動風險", 'label2'] = 3
df_news.loc[df_news['label'] == "管理風險", 'label2'] = 4
df_news['label2'] = df_news['label2'].astype(int)
df_news
```

```python
from snownlp import SnowNLP

# text rank 演算法 死出top5的關鍵字
def keywords_all(text):
    s = SnowNLP(u""+text)       # 宣告物件 s = SnowNLP(u"这个东西真心很不错呀")
    text2 = s.han               # snownlp 要簡體中文效果比較好
    s = SnowNLP(u""+text2)
    return s.keywords(5)

def sentiments_all(text):
    s = SnowNLP(u""+text)       # 宣告物件 s = SnowNLP(u"这个东西真心很不错呀")
    text2 = s.han               # snownlp 要簡體中文效果比較好
    s = SnowNLP(u""+text2)
    return s.sentiments
```

```python
df_news['keywords'] = df_news['text'].apply(keywords_all)
df_news['sentiments'] = df_news['text'].apply(sentiments_all)
df_news
```
